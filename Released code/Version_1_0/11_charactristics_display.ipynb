```python
temp=[0 for i in range(24)]
temp_index=[0 for i in range(24)]
pres=[0 for i in range(24)]
pres_index=[0 for i in range(24)]
humi=[0 for i in range(24)]
humi_index=[0 for i in range(24)]
```

```python
# initial PM2.5 PM10 NO2 list for record data into a day
pm25_day=[0 for i in range(24)]
pm25_day_index=[0 for i in range(24)]
pm10_day=[0 for i in range(24)]
pm10_day_index=[0 for i in range(24)]
no2_day=[0 for i in range(24)]
no2_day_index=[0 for i in range(24)]
```

```python
# Calculate the average value of PM2.5, PM10, NO2
for x in station_data:
    for index, row in station_data[x].iterrows():
        if row["PM25_Concentration"] != 0:
            pm25_day[int(row["time_hour"]%24)] += row["PM25_Concentration"]
            pm25_day_index[int(row["time_hour"]%24)] += 1
        if row["PM10_Concentration"] != 0:
            pm10_day[int(row["time_hour"]%24)] += row["PM10_Concentration"]
            pm10_day_index[int(row["time_hour"]%24)] += 1
        if row["NO2_Concentration"] != 0:
            no2_day[int(row["time_hour"]%24)] += row["NO2_Concentration"]
            no2_day_index[int(row["time_hour"]%24)] +=1
```

```python
for i in range(24):
    pm25_day[i] /= pm25_day_index[i]
    pm10_day[i] /= pm10_day_index[i]
    no2_day[i] /= no2_day_index[i]
```

```python
# Temperature, Pressure, and Humidity
for x in station_data:
    for index, row in station_data[x].iterrows():
        if row["temperature"] != 0:
            temp[int(row["time_hour"]%24)] += row["temperature"]
            temp_index[int(row["time_hour"]%24)] += 1
        if row["pressure"] != 0:
            pres[int(row["time_hour"]%24)] += row["pressure"]
            pres_index[int(row["time_hour"]%24)] += 1
        if row["humidity"] != 0:
            humi[int(row["time_hour"]%24)] += row["humidity"]
            humi_index[int(row["time_hour"]%24)] +=1
```

```python
#temp,temp_index,pres,pres_index,humi,humi_index
```

```python
for i in range(24):
    temp[i] = temp[i] / temp_index[i]
    pres[i] = pres[i] / pres_index[i]
    humi[i] = humi[i] / humi_index[i]
```

```python
# Month phase

# Records
pm25_mon=[0 for i in range(12)]
pm25_mon_index=[0 for i in range(12)]
pm10_mon=[0 for i in range(12)]
pm10_mon_index=[0 for i in range(12)]
no2_mon=[0 for i in range(12)]
no2_mon_index=[0 for i in range(12)]
```

```python
# Calculate the average value of PM2.5, PM10, NO2
for x in station_data:
    for index, row in station_data[x].iterrows():
        if row["PM25_Concentration"] != 0:
            pm25_mon[int(row["time_month"]%12 - 1)] += row["PM25_Concentration"]
            pm25_mon_index[int(row["time_month"]%12 - 1)] += 1
        if row["PM10_Concentration"] != 0:
            pm10_mon[int(row["time_month"]%12 - 1)] += row["PM10_Concentration"]
            pm10_mon_index[int(row["time_month"]%12 - 1)] += 1
        if row["NO2_Concentration"] != 0:
            no2_mon[int(row["time_month"]%12 - 1)] += row["NO2_Concentration"]
            no2_mon_index[int(row["time_month"]%12 - 1)] +=1
```

```python
# Average in month
for i in range(12):
    pm25_mon[i] /= pm25_mon_index[i]
    pm10_mon[i] /= pm10_mon_index[i]
    no2_mon[i] /= no2_mon_index[i]
```

```python
fig = plt.figure(figsize=(12,5))

ax1 = fig.add_subplot(111)
ax1.plot(range(1,13), pm25_mon, ".-", label="PM2.5")
ax1.plot(range(1,13), pm10_mon, ".-", label="PM10")
ax1.plot(range(1,13), no2_mon, ".-", label="NO2")
ax1.set_xlabel("Month")
ax1.set_ylabel(r"Pollutant Concentration (ug/m$^3$)")
ax1.set_title("Pollutant Concentration relationship with Month")
plt.legend(loc=0)
plt.show()
```
