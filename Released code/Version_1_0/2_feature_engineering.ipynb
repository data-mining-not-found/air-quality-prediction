```python
import requests
import pandas as pd
import numpy as np
import pickle
```

```python
def save_pkl(file_dir, data):
    f = open(file_dir,"wb")
    pickle.dump(data, f, protocol=4)
    f.close()
    
def read_pkl(file_dir):
    f = open(file_dir,"rb")
    data = pickle.load(f)
    return data
```

```python
# merge the 64 weather forecasting features

def add_increasing_random(original_matrix):
    result_matrix = np.dot( original_matrix , np.diag(1 + np.multiply(np.arange(1,1.24,0.005) , (np.random.random_sample((48,)) - 0.5) / 5)))
    return np.around(result_matrix)

def merge_weather_forecast (station_group, length = 10805):
    for name in station_group:
        print(name)
        temperature_list = station_group[name]['temperature'].values
        pressure_list = station_group[name]['pressure'].values
        humidity_list = station_group[name]['humidity'].values
        wind_direction_list = station_group[name]['wind_direction'].values
        original_temperature_list = []
        original_pressure_list = []
        original_humidity_list = []
        original_wind_direction_list = []
        
        l = len(station_group[name])
        extra_array = np.zeros(l - length - 1)
        
        for i in range(length+1):
            original_temperature_list.append(temperature_list[i+1 : i+49])
            original_pressure_list.append(pressure_list[i+1 : i+49])
            original_humidity_list.append(humidity_list[i+1 : i+49])
            original_wind_direction_list.append(wind_direction_list[i+1 : i+49])
            
        original_temperature = np.vstack(original_temperature_list)
        original_pressure = np.vstack(original_pressure_list)
        original_humidity = np.vstack(original_humidity_list)
        original_wind_direction = np.vstack(original_wind_direction_list)
        
        processed_temperature = add_increasing_random (original_temperature)
        processed_pressure = add_increasing_random (original_pressure)
        processed_humidity = add_increasing_random (original_humidity)
        processed_wind_direction = add_increasing_random (original_wind_direction)
        
        for i in range(16):
            sample_temperature = processed_temperature[:,i*3+2]
            feature_name = 'te_' + str(i)
            station_group[name][feature_name] = np.append(sample_temperature, extra_array)
        
        for i in range(16):
            sample_pressure = processed_pressure[:,i*3+2]
            feature_name = 'pr_' + str(i)
            station_group[name][feature_name] = np.append(sample_pressure, extra_array)
        
        for i in range(16):
            sample_humidity = processed_humidity[:,i*3+2]
            feature_name = 'hu_' + str(i)
            station_group[name][feature_name] = np.append(sample_humidity, extra_array)
        
        for i in range(16):
            sample_wind_direction = processed_wind_direction[:,i*3+2]
            feature_name = 'wd_' + str(i)
            station_group[name][feature_name] = np.append(sample_wind_direction, extra_array)
        
#     print(original_temperature[:100])    
#     print(processed_temperature[:100])  
    
merge_weather_forecast(station_group)
station_group['ST5']
```
